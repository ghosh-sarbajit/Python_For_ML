```python
# Replacing missing value with mean

df = pd.read_csv('data.csv')
# print(df.to_string())
x = df["Calories"].mean()
# print(df.to_string())
# x = df["Calories"].mode()[0]
print(x)
df["Calories"].fillna(x, inplace = True)
print(df.to_string())

# replace duration
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120
    
# Remove rows having some outliers
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.drop(x, inplace = True)
    
df.corr()

```

```python
# Correct formatting

df = pd.read_csv('dirtydata.csv')
df['Date'] = pd.to_datetime(df['Date'])
# print(df.to_string())

# Rm rows with a NULL value in the "Date" col
df.dropna(subset=['Date'], inplace = True)
# print(df.to_string())

# print(df.duplicated())
df.drop_duplicates(inplace = True)
print(df.to_string())

```

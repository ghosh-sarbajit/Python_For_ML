```python
# Liner reg
# import matplotlib.pyplot as plt
# from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

# ore 'r' ekhane correlation Not R^2

slope, intercept, r, p, std_err = stats.linregress(x, y)

print(slope, "\n",intercept, "\n", r, "\n", p, "\n", std_err)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
from sklearn.metrics import r2_score
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]


# mymodel = numpy.polyfit(x, y, 3)
# print(mymodel)
mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))
print(mymodel)

# takes 100 points between 1 to 22
myline = numpy.linspace(1, 22, 100)
# print(myline)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()

# Print R^2
print(r2_score(y, mymodel(x)))

# Future prediction
speed = mymodel(17)
print(speed)
```

```python
# Making train and test set
import numpy
import matplotlib.pyplot as plt
numpy.random.seed(2)
from sklearn.metrics import r2_score


x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(350, 40, 100)
# print(x)
# print(y)
y=y/x
# print(y)

# x = numpy.array(range(10))
# x=3*x
# # print(x)
# y = numpy.array(range(10))
# z = x[1:]/y[1:]
# print(z)
plt.scatter(x, y)
plt.show()

# Train data plotting
train_x = x[:80]
train_y = y[:80]

plt.scatter(train_x, train_y)
plt.show()

# Test data plotting
test_x = x[80:]
test_y = y[80:]

plt.scatter(test_x, test_y)
plt.show()

# Fitting a model
mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 2))

myline = numpy.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()

r2 = r2_score(train_y, mymodel(train_x))
print(r2)

r2 = r2_score(test_y, mymodel(test_x))
print(r2)

print(mymodel(5))
```

```python
import pandas
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg

df = pandas.read_csv("shows.csv")

print(df)
```

```python
d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

print(df)
```

```python
features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']

# print(X)
# print(y)
```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)
data = tree.export_graphviz(dtree, out_file=None, feature_names=features)
graph = pydotplus.graph_from_dot_data(data)
graph.write_png('mydecisiontree.png')

img=pltimg.imread('mydecisiontree.png')
imgplot = plt.imshow(img)
plt.show()
```
